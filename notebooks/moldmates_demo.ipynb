```python
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
from moldmates.constants import LINES
from moldmates.load import load_dir
from moldmates.objects import Image
from moldmates.transform import BestAligner

# this is the function that parses all the data files
# it also does some preprocessing like mean-centering and normalizing
image_set = load_dir(LINES)
```

```python
# intiialize figure
plt.figure(figsize=(20, 10))
ax = plt.gca()

# get two images to align
image1 = image_set['Linescodex_01A_backlit.txt']
image2 = image_set['Linescodex_01B_backlit.txt']

# initialize the "aligner" on the image we are aligning TO
# n_chainlines is how many chainlines to consider in this alignment
aligner = BestAligner(n_chainlines=13).fit(image1)

# plot initial image
image1.plot(ax)

# find the best transformation of the second image and plot it
aligner.transform(image2).plot(ax)
```
